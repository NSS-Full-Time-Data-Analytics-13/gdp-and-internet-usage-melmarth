```python
#Which country was the first to have a GDP per capita greater than $100,000?
gdp_df.loc[gdp_df['GDP_per_Capita'] >= 100000].sort_values(['Year','GDP_per_Capita'], ascending = [True,False]).head(1)
```

```python
#Which country had the highest GDP per capita in 2020? 
#Create a plot showing how this country's GDP per capita has changed over the timespan of the dataset.
max_gdp_2020 = gdp_df['GDP_per_Capita'].loc[gdp_df['Year'] == 2020].max()
max_2020 = gdp_df.loc[gdp_df['GDP_per_Capita'] == max_gdp_2020]
print(max_2020.iloc[0,0] + ' had the highest GDP per capita in 2020')
```

```python
sns.scatterplot(x = gdp_df['Year'], y = gdp_df['GDP_per_Capita'].loc[gdp_df['Country'] == max_2020.iloc[0,0]]).set(title = max_2020.iloc[0,0]);
```

```python
#Which country had the lowest GDP per capita in 2020? 
min_gdp_2020 = gdp_df['GDP_per_Capita'].loc[gdp_df['Year'] == 2020].min()
min_2020 = gdp_df.loc[gdp_df['GDP_per_Capita'] == min_gdp_2020]
print(min_2020.iloc[0,0] + ' had the lowest GDP per capita in 2020')
#Create a plot showing how this country's GDP per capita has changed over the timespan of the dataset. 
sns.scatterplot(x = gdp_df['Year'], y = gdp_df['GDP_per_Capita'].loc[gdp_df['Country'] == min_2020.iloc[0,0]]).set(title = min_2020.iloc[0,0]);
```

```python
#Bonus question: Is it true in general that coutries had a higher GDP per capita in 2020 than in 1990? 
#Which countries had lower GDP per capita in 2020 than in 1990?
gdp_df_diff = gdp_df.loc[gdp_df['Year'] == 2020].rename(columns = {'GDP_per_Capita': 'gdp_2020'}).drop(columns = 'Year')
gdp_df_diff['gdp_1990'] = 0
for i, row in gdp_df.iterrows():
    if row.Year == 1990:
        gdp_df_diff['gdp_1990'].loc[gdp_df_diff['Country'] == row.Country] = row.GDP_per_Capita

gdp_df_diff['diff'] = gdp_df_diff['gdp_2020'] - gdp_df_diff['gdp_1990']
gdp_df_diff.head()
```

```python
print('Most countries had a higher GDP in 2020 than 1990.')
sns.boxplot(x = gdp_df_diff['diff']);
```

#### Read in the internet use dataset into a DataFrame

```python
internet_df = pd.read_csv('../data/internet_use.csv', usecols=['Country or Area', 'Year', 'Value'], nrows = 6083).rename(columns = {'Value' : 'Percent_Use', 'Country or Area' : 'Country'})
```

```python
internet_df
```

```python
#number of observations in this dataset per year - missing data
internet_df['Year'].value_counts()
```

```python
#What is the first year to have a non-zero internet users percentage value
for index, row in internet_df.sort_values('Year').iterrows():
    if row['Percent_Use'] > 0:
        print(row.Year)
        break
```

```python
#How does the distribution of internet users percent differ for 2000 and 2014?
sns.boxplot(x = internet_df['Year'].loc[internet_df['Year'].isin([2000, 2014])], y = internet_df['Percent_Use'])
```

```python
#For how many countries was the percentage of internet users below 5% in 2014?
internet_df.loc[(internet_df['Percent_Use'] < 5) & (internet_df['Year'] == 2014)].shape[0]
```

```python
#Merge the two DataFrames to one. Do this in a way that keeps all rows from each of the two DataFrames. Call the new DataFrame gdp_and_internet_use.
gdp_and_internet_use = pd.merge(gdp_df, internet_df, how = 'inner', on = ['Country', 'Year'])

#Look at the first and last few rows to confirm that it merged correctly
gdp_and_internet_use
```

```python
#Find the three countries with the highest internet users percentage in 2014. 
top_3 = gdp_and_internet_use.loc[gdp_and_internet_use['Year'] == 2014].nlargest(3,'Percent_Use')
top_3['Country']
```

```python
#Use a seaborn FacetGrid to compare how the GDP per capita has changed over time for these three countries.
top_3_time = gdp_and_internet_use.loc[gdp_and_internet_use['Country'].isin(top_3['Country'])]
g = sns.FacetGrid(top_3_time, col='Country')
g.map_dataframe(sns.scatterplot, x='Year', y='Percent_Use')
g.add_legend();
```

```python
#Subset gdp_and_internet_use to just the year 2014. Save this as a new dataframe named gdp_and_internet_use_2014
gdp_and_internet_use_2014 = gdp_and_internet_use.loc[gdp_and_internet_use['Year'] == 2014]
gdp_and_internet_use_2014
```

```python
#Create a plot which compares Internet Users Percentage and GDP per Capita for the year 2014.
sns.scatterplot(gdp_and_internet_use_2014, x = 'GDP_per_Capita', y = 'Percent_Use').set(title = 'GDP vs Internet 2014');
```

```python
#Use the qcut function from pandas to divide countries in gdp_per_capita_2014 into three groups based on their GDP per capita values. 
#Label these groups as "Low", "Medium", and "High". 
gdp_cat = pd.qcut(gdp_and_internet_use_2014['GDP_per_Capita'], 3, labels=["Low", "Medium", "High"]).tolist()
```

```python
#Put these labels in a new column, named "GDP_group".
gdp_and_internet_use_2014['GDP_group'] = gdp_cat
gdp_and_internet_use_2014
```

```python
sns.scatterplot(gdp_and_internet_use_2014, x = 'GDP_per_Capita', y = 'Percent_Use', hue = 'GDP_group').set(title = 'Internet Use and GDP');
```

```python
#How does the median internet users percentage compare for the three gdp groups?
sns.boxplot(gdp_and_internet_use_2014, x = 'GDP_group', y = 'Percent_Use').set(title = 'Internet Use Among GDP Groups');
```
